```python
import requests
from bs4 import BeautifulSoup

url = "https://kazan.cian.ru/snyat-sklad/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
t = soup.find_all('li', class_="_32bbee5fda--factoid-item--ABc2c")
gg = soup.find_all('div',class_="_32bbee5fda--userInfo-name--atM0D")
a=[]
print(len(gg))
gg_s=[]
for i in range(1, len(list(t)), 2):
  r = str(t[i]).split('</li>')
  rr= r[0].split('data-name="CommercialFactoid">')
  a.append(rr[1])

for il in range(1, len(list(gg)), 2):
  ry = str(gg[il]).split('</noindex></div>')
  rry= ry[0].split('<noindex>')
  gg_s.append(rry[1])

f=[]
b = a.copy()
for k in range(0, len(b)):
  b[k] = b[k]. split(' руб.')
  f.append(b[k][0])
for y in range(len(f)):
  f[y]=f[y].replace(' ', '')
  f[y]=int(f[y])
dicdic={'арендодатель': gg_s, 'красивое': a[:13], 'цена': f[:13]}

dfgh=pd.DataFrame(dicdic)
dfgh = dfgh.sort_values(by=['цена'])
dfgh= dfgh.drop([1])
dfgh = dfgh.reset_index(drop=True)
dfgh
```

```python
import dash
from dash import Dash, dcc, html, Input, Output
import dash_bootstrap_components as dbc
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

resultryu = pd.concat([df_mint, df_ivan, df_melissa])

app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

def calculate_statistics(species_filter):
    if species_filter == 'мята':
        df = df_mint
    elif species_filter == 'иван-чай':
        df = df_ivan
    elif species_filter == 'мелисса':
        df = df_melissa
    elif species_filter == 'Сам ты пень, а я склад':
        df = dfgh
    else:
        df = resultryu

    stats = {
        'количество продавцов': len(df),
        'средняя цена': round(df['цена'].mean(), 2),
    }
    return stats

def calculate_top(species_filter):
    if species_filter == 'Сам ты пень, а я склад':
        df = dfgh
        top = {
            'first': df.iloc[0]['арендодатель'],
            'second': df.iloc[1]['арендодатель'],
            'third': df.iloc[2]['арендодатель']
        }
    else:
        if species_filter == 'мята':
            df = df_mint
        elif species_filter == 'иван-чай':
            df = df_ivan
        elif species_filter == 'мелисса':
            df = df_melissa
        else:
            df = resultryu

        top = {
            'first': df.iloc[0]['город'],
            'second': df.iloc[1]['город'],
            'third': df.iloc[2]['город']
        }
    return top

def generate_graph(species_filter):
    if species_filter == 'Сам ты пень, а я склад':
        fig = px.histogram(dfgh, x='цена', hover_data=['арендодатель', 'красивое'], nbins=20)
        fig.update_layout(
            height=500, width=1200,
            title='prices in kazan',
            xaxis_title='prices',
            yaxis_title='Частота',
            legend_title='no idea',
            xaxis=dict(range=[1000, 30000])
        )
        fig.update_traces(
            marker=dict(color='green', line=dict(color='darkgreen', width=1)),
            selector=dict(type='histogram')
        )
        fig_price = px.histogram(dfgh, x='цена', title='Распределение цены')
        fig_price.update_traces(
            marker=dict(color='green', line=dict(color='darkgreen', width=1)),
            selector=dict(type='histogram')
        )
    else:
        if species_filter == 'мята':
            df = df_mint
            title = 'Мята'
            symbol = 'star'
        elif species_filter == 'иван-чай':
            df = df_ivan
            title = 'Иван чай'
            symbol = 'circle'
        elif species_filter == 'мелисса':
            df = df_melissa
            title = 'Мелисса'
            symbol = 'cross'
        else:
            df = resultryu
            title = 'Все растения'
            symbol = 'circle'

        fig = px.scatter_geo(df, lat='широта', lon='долгота', hover_name='город',
                             color='обратная_цена', size='обратная_цена',
                             projection='equirectangular', title=title, symbol_sequence=[symbol],
                             color_continuous_scale='RdYlGn', text='город', width=1200,
                             height=700, opacity=0.8)
        fig_price = px.histogram(df, x='цена', title='Распределение цены')
        fig_price.update_traces(
            marker=dict(color='green', line=dict(color='darkgreen', width=1)),
            selector=dict(type='histogram')
        )
    return fig, fig_price

app.layout = dbc.Container([
    html.H1('Чайный дашборд', className='text-center text-success mb-4'),
    dbc.Row([
        dbc.Col([
            dbc.Card([
                dbc.CardHeader('Фильтр по видам растений'),
                dbc.CardBody([
                    dcc.Dropdown(
                        id='species-dropdown',
                        options=[
                            {'label': 'Мята', 'value': 'мята'},
                            {'label': 'Иван чай', 'value': 'иван-чай'},
                            {'label': 'Мелисса', 'value': 'мелисса'},
                            {'label': 'Сам ты пень, а я склад', 'value': 'Сам ты пень, а я склад'}
                        ],
                        value='мята',
                        placeholder='Выберите растение'
                    )
                ])
            ])
        ], md=4),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader('Основные статистики'),
                dbc.CardBody([
                    html.P(id='количество продавцов', className='card-text'),
                    html.P(id='средняя цена', className='card-text'),
                ])
            ])
        ], md=8)
    ], className='mb-4'),
    dbc.Row([
        dbc.Col(dcc.Graph(id='geo-graph'), md=12)
    ]),
    dbc.Row([
        dbc.Col([dbc.CardHeader('Распределение цен'),
                dbc.CardBody([
                    dcc.Graph(id='price_dist')
                ]) ], md=6),
        dbc.Col([
            dbc.CardHeader('Три города с самыми маленькими ценами'),
            dbc.CardBody([
                html.P(id='first', className='card-text'),
                html.P(id='second', className='card-text'),
                html.P(id='third', className='card-text')
            ])
        ], md=6),
    ]),
], fluid=True)

@app.callback(
    [Output('количество продавцов', 'children'),
     Output('средняя цена', 'children'),
     Output('geo-graph', 'figure'),
     Output('price_dist', 'figure'),
     Output('first', 'children'),
     Output('second', 'children'),
     Output('third', 'children')],
    [Input('species-dropdown', 'value')]
)
def update_dashboard(species_filter):
    try:
        print(f"Filter: {species_filter}")
        stats = calculate_statistics(species_filter)
        print("Stats:", stats)
        fig, fig_price = generate_graph(species_filter)
        print("Graphs generated successfully")
        top = calculate_top(species_filter)
        print("Top cities:", top)

        measurement_count = f"количество продавцов: {stats['количество продавцов']}"
        avg_petal_length = f"средняя цена: {stats['средняя цена']} руб."
        first = f"1. {top['first']}"
        second = f"2. {top['second']}"
        third = f"3. {top['third']}"

        return measurement_count, avg_petal_length, fig, fig_price, first, second, third
    except Exception as e:
        print(f"Error in callback: {e}")
        return "Error", "Error", go.Figure(), go.Figure(), "N/A", "N/A", "N/A"

if __name__ == '__main__':
    app.run(debug=True)
```
